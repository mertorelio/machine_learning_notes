### Değişken Standardizasyonu

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
            {"V1" : V1,
             "V2" : V2,
             "V3" : V3
            }
)
df = df.astype(float)
df
```

#### Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df) #yapışarı birbiriyle kıyaslanabilirliği bozulmadı
#değişkenleri standartlaştırıp algoritmayı yormayız
```

#### Normalizasyon
Değişkenleri 0-1 arasına almak için kullanılır.

```python
preprocessing.normalize(df)
```

#### Min-Max Dönüşümleri
Kendi istediğimiz aralıklara çekeriz değişkenleri

```python
scaler =preprocessing.MinMaxScaler(feature_range=(10,20))
```

```python
scaler.fit_transform(df)
```

### Değişken Dönüşümleri
Her standartliazasyon bir değişken dönüşümüdür. Önemli olan değişkenin yapısın bozulup bozulmamasıdır

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

##### 0-1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
```

```python
df["yeni_sex"] = lbe.fit_transform(df['sex']) #cinsiyet değişkenini 1 ve 0 a dönüştürdük
```

```python
df #veriye göre neye 1 neye 0 verdiğine dikkat edelim
```

##### "1 ve Diğerleri (0) Dönüşümü

```python
df.head()
```

```python
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)
#içinde sun olan str şeklindeki değerler yerine 1 diğerlerine de 0 yaz dedik
```

```python
df
```

#### Çok Sınıflı Dönüşüm

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["day"]) #day içindeki sınıflar numerik değerlere dönüştü
#fakat günleri sayıya çevirdiğimiz için günler değei olmuş olur
```

#### One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df,columns = ["sex"],prefix =["sex"])
```

```python
df_one_hot.head()
#cinsiyetleri kategorik olarak ayırıp 1 ve 0 a dönüştürüyoruz bune one hot denir
```
