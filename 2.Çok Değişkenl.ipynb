### Çok Değişkenli Aykırı Gözlem Analizi

#### Local Outlier Factor
Bir nokta etrafındaki diğer komşularına göre yoğunluğu daha seyrek ise bu durum aykırı bir durum olarak ele alınır

```python
import seaborn as sns
df = sns.load_dataset('diamonds') #veri setini çekiyoruz
df = df.select_dtypes(include = ['float64','int64']) #veri setinden sadece belli değerleri çekmek için
df = df.dropna() #eksik değerleri bırakmayı sağlar
df.head()
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)
```

```python
clf.fit_predict(df)
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:10] #yoğunluk skoru
```

```python
np.sort(df_scores)[0:20]#sıraladık ve ilk 20ye baktık
```

```python
esik_deger = np.sort(df_scores)[13]#esik değer belirledik
```

```python
aykiri_tf = df_scores > esik_deger
```

```python
yeni_df = df[df_scores> esik_deger]
```

```python
yeni_df #esik değerin üzerindekiler bunlar sildik aykırı değer yok burda bi nevi silme işlemi yapmış olduk
```

```python
df[df_scores < esik_deger] #bunlar da aykırı değerlerdir
```

### Baskılama işlemi

```python
df[df_scores ==esik_deger] #eşik dğerimiz bunlar
```

```python
baski_deger = df[df_scores ==esik_deger]
```

```python
aykirilar = df[~aykiri_tf]
```

```python
res = aykirilar.to_records(index = False) #aşağıdaki bir dizi işlemler bazı index problerini giderebilme adına yapılmıştır
#indexsiz bir değer dönüştürüp daha sonra tekrar düzelttik
```

```python
res[:] = baski_deger.to_records(index = False)
```

```python
res
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index)
```
